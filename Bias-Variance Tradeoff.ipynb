```python
# Bias-Variance Tradeoff
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#data 
X = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)
y = np.array([3,6,7,9,12,15,16,18])
```

```python
# high bias model (Underlifitting)
model_bias = LinearRegression()
model_bias.fit(X,y)
```

```python
# high variance model(overfitting)
model_variance = Pipeline([
    ('poly',PolynomialFeatures(degree=6)),
    ('LR',LinearRegression())
])
model_variance.fit(X,y)
```

```python
#plot
plt.scatter(X,y)
plt.plot(X,model_bias.predict(X),label='High Bias',linestyle ='--')
plt.plot(X,model_variance.predict(X),label='High Variance')
plt.legend()
plt.show()
```
