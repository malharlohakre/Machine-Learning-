```python
# svm regression aur classification dono krdeti dono kaam sathmai toh regression krne ki zarut nhi sirf classfication krna hai toh random forest
#svm finds a best boundary that seprates classes with maximum margin
# RBF - Radial Basis Function
# SVM - Support Vector Machines
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
X = iris.data[:,:2] # first two features for 2D
y = iris.target
#print(y)
```

```python
# train test
X_train, x_test ,y_train, y_test = train_test_split(X,y,test_size=0.2) # 20% data in testing can also increse if want 
# also make 4 variable X_train, x_test ,y_train, y_test

#rbf kernel SVM
model = svm.SVC(kernel='rbf',gamma=0.5) # kernel='rbf' always write
model.fit(X_train,y_train)
```

```python
# RBF kernel SVM with accuracy
accuracy = model.score(X_train,y_train)
#plot decision boundary  setting boundary is compulsory

x_min, x_max = X[:,0].min()-1,X[:,0].max()+1   # use 0 result will be better
y_min, y_max = X[:,1].min()-1,X[:,1].max()+1 

xx, yy = np.meshgrid(
    np.linspace(x_min,x_max,300),
    np.linspace(y_min,y_max,300)
)

Z = model.predict(np.c_[xx.ravel(),yy.ravel()]) #ravel its a function
Z = Z.reshape(xx.shape)
plt.figure(figsize=(8,6))

#decision boundary
plt.contourf(xx,yy,Z,alpha=0.3)
#scatter plot
plt.scatter(X[:,0],X[:,1],c=y) # c=y is predition 
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('SVM with RBF')
plt.show()
```
