```python
#multiple linear regression 21/11/25
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
```

```python
data = {
    'Size':[1000,1500,2000,1200,1800],
    'Bedrooms':[2,3,3,2,4],
    'Price':[150,220,280,170,260]
}
```

```python
df = pd.DataFrame(data)
x = df[['Size','Bedrooms']]
y = df['Price']
```

```python
#step 2 fit the model
model = LinearRegression()
model.fit(x,y)
print("Intercept:(Beta 0)",model.intercept_)
print("Coefficients: (B1,B2)",model.coef_)
```

```python
#predict the value 1600 sqft and 3 bedrooms
new_house = np.array([[1600,3]])
predicted_price = model.predict(new_house)
print(f"Predicted price for 16000 sqft and 3 bedrooms:,{predicted_price[0]:.2F}")
```

```python
# step 4 evaluate the model
y_pred = model.predict(x)
mse = mean_squared_error(y,y_pred)
print(f"Mean squared error:{mse:.2F}")
```

```python
r2 = r2_score(y,y_pred)
print(f"R-squared: {r2:.2F}")
```
