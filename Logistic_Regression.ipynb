```python
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.metrics import confusion_matrix , precision_score, recall_score, f1_score
```

```python
hours = np.array([[2],[4],[6]])
answers = np.array([0,1,1]) # 0-fail , 1-pass

model = LogisticRegression()
model.fit(hours,answers)

#learned by model
print("starting point",model.intercept_[0])
print("coeefficient",model.coef_[0][0])

#make guesses
guess = model.predict(hours)
print("guesses",guess)
```
