```python
import numpy as np
from sklearn.linear_model import LinearRegression # remember this line always in linear regression 
import matplotlib.pyplot as plt
```

```python
#data
x = np.array([8,10,12]).reshape(-1,1) # pizza diameter (independent variable)
y = np.array([10,13,16])# pizza price (dependent variable)
```

```python
# create and train the model
# Y = mX + b
model = LinearRegression()
model.fit(x,y)
```

```python
# get the slope (m)
# get the intercept(b)
m = model.coef_[0] # slope # function of removing slope remember it 
b = model.intercept_ # Intercept
print(f"slope:{m}")
print(f"slope:{m:.2f}") #formatting getting a proper number
print(f"Intercept:{b}")
```

```python
# predict price for 20 inch = prize ?
x_new = np.array([20]).reshape(-1,1)
y_predict = model.predict(x_new) #predict this function is used for predictions
print(f"Predict price for 20 inch pizza = {y_predict[0]:.2f}")
```

```python
#plot the data point
plt.scatter(x,y,color='blue',label='data point')
plt.plot(x,model.predict(x),color='green',label='regression line')
plt.xlabel('diameter (inch)')
plt.ylabel("price $")
plt.title("pizza price predictions")
plt.legend()
plt.show()
```
